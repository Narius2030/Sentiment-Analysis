```python
# -*- coding: utf-8 -*-
from gensim.models import Word2Vec
import os
import pandas as pd
```

```python
path = './data/sentiments'
def read_data(path):
    list_file = os.listdir(path)
    print(list_file)
    data = pd.DataFrame()
    for filename in list_file:
      if 'xlsx' not in filename:
        data = pd.concat([data, pd.read_csv(os.path.join(path, filename), sep = ',', encoding='utf-8')])

    return data.Review, data.Label
```
